```python
from sklearn import datasets
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/npradaschnor/Pima-Indians-Diabetes-Dataset/master/diabetes.csv')
df.shape
```

```python
df.sample(3)
```

```python
X=df.drop(columns=['Outcome'])
Y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
numeric_col=X_train.columns
numeric_col
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

transformer=ColumnTransformer(transformers=[
    ('scaler',StandardScaler(),numeric_col)
],remainder='passthrough')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline

pipeline=make_pipeline(transformer,RandomForestClassifier())
pipeline.fit(X_train,Y_train)
```

```python
rf=pipeline.named_steps['randomforestclassifier']
print(rf.feature_importances_)
```

```python
print(pipeline.named_steps.keys())
```

```python
import matplotlib.pylab as plt
import numpy as np

pd.Series(np.abs(rf.feature_importances_),index=X_train.columns).sort_values(ascending=False).plot(kind='bar')
plt.show()
```

# SelectFromModel

```python
from sklearn.feature_selection import SelectFromModel

model=RandomForestClassifier()
```

```python
pipeline=make_pipeline(transformer,SelectFromModel(model,threshold='mean'))
pipeline.fit(X_train,Y_train)
```

```python
sfm=pipeline.named_steps['selectfrommodel']
print(sfm.estimator_.feature_importances_)
```

```python
pd.Series(sfm.estimator_.feature_importances_,index=X.columns).sort_values(ascending=False).plot(kind='bar')
plt.show()
```

**Important Points:**
1. pipeline.named_steps['randomforestclassifier']:
This line accesses the RandomForestClassifier step inside your pipeline. When you use make_pipeline, scikit-learn automatically names each step after the lowercase version of the class name (e.g., 'randomforestclassifier').

2. rf.feature_importances_:
This attribute gives you the feature importances learned by the random forest model after fitting. These values indicate how much each input feature contributed to the model’s predictions.
