```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("train.csv")
df.shape
```

```python
df.head()
```

```python
X=df.drop(columns=['TARGET'])
Y=df['TARGET']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import SelectKBest
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
pipeline=make_pipeline(
    StandardScaler(),
    VarianceThreshold(),
    SelectKBest(f_classif,k=200),
    RandomForestClassifier()
)
```

```python
pipeline.fit(X_train,Y_train)
```

```python
from sklearn.metrics import f1_score

Y_pred=pipeline.predict(X_test)
print(f"The f1 score is: {f1_score(Y_pred,Y_test)}")
```

**Important Points:**
1. ANOVA is univerate, meaning it just looks at how one feature is realted to the target vairable. Sometimes two features combined might result a better output which ANOVA completely neglects. Thus, use ANOVA only when you have a lot of columns and you want to drop them quickly.
2. It's a good practise to use Variance Threshold before ANOVA as constant terms might create problem while using ANOVA.
