```python
import pandas as pd
import matplotlib.pylab as plt
plt.style.use('fivethirtyeight')
import seaborn as sns
```

```python
df=pd.read_csv("train.csv",usecols=['GarageQual','FireplaceQu','SalePrice'])
df.sample(3)
```

```python
df['FireplaceQu'].value_counts().plot(kind='bar')
plt.show()
```

```python
df['GarageQual'].value_counts().plot(kind='bar')
plt.show()
```

```python
df.isnull().mean()*100
```

```python
ta_prices=df[df['GarageQual']=='TA']['SalePrice']
null_prices=df[df['GarageQual'].isnull()]['SalePrice']
```

```python
#GarageQual Feature
fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(10,6))

sns.kdeplot(data=ta_prices,label='Houses with TA',fill=True)
sns.kdeplot(data=null_prices,label='Houses with NA',fill=True)

plt.legend()
plt.show()
```

```python
df['GarageQual'].fillna("TA",inplace=True)
```

```python
df['GarageQual'].value_counts().plot(kind='bar')
plt.show()
```

```python
mode_imputated=df[df['GarageQual']=='TA']['SalePrice']
```

```python
fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(10,6))

sns.kdeplot(data=ta_prices,label='Houses with TA',fill=True)
sns.kdeplot(data=mode_imputated,label='Houses after mode imputation',fill=True)

plt.legend()
plt.show()
```

```python
df['FireplaceQu'].value_counts().plot(kind='bar')
plt.show()
```

```python
fireplace=df[df['FireplaceQu']=='Gd']['SalePrice']
```

```python
fireplace_null=df[df['FireplaceQu'].isnull()]['SalePrice']
```

```python
#FireplaceQu Feature
fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(10,6))

sns.kdeplot(data=fireplace,label='Houses with GD',fill=True)
sns.kdeplot(data=fireplace_null,label='Houses with NA',fill=True)

plt.legend()
plt.show()
```

```python
new_df=df.copy()
```

```python
df['FireplaceQu'].fillna('Gd',inplace=True)
```

```python
df['FireplaceQu'].value_counts().plot(kind='bar')
plt.show()
```

```python
fireplace_imputed=df[df['FireplaceQu']=='Gd']['SalePrice']
```

```python
#FireplaceQu Feature
fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(10,6))

sns.kdeplot(data=fireplace,label='Houses with GD',fill=True)
sns.kdeplot(data=fireplace_imputed,label='Imputed',fill=True)

plt.legend()
plt.show()
```

Using a missing category

```python
new_df['FireplaceQu'].fillna('Missing',inplace=True)
```

```python
new_df['FireplaceQu'].value_counts().plot(kind='bar')
plt.show()
```

**Important Points**
1. For mode imputation in categorical columns it's a good practise if it satisfies two conditions:
   a. There is less that 5% missing value.
   b.  If one category (e.g., TA for GarageQual) significantly outweighs others, mode imputation is reasonable because you're not introducing much       distortion.
2. Here, we imputed with Garage Quality with the mode of that features. It's more beneficial if we can draw a kde plot wrt to the target variable in categorical variables to extract more information. Our first graph is a overlapped kde plot of houses prices with typical/average(mode) and house prices of the missing values. The graph gives a rough estimation that the prices of houses with missing value is less indicating that the garage quality might be below typical. The second plot is a prices distribution of typical garage quality before and after mode imputation. Since, imputing by mean does not have any effect on the original distribution of price of typical category, it's safe to use it.
   
Insight: There might be a systematic biasmissing values are not missing completely at random but perhaps associated with lower-quality garages (which might have been left unrecorded).

3. Imputing the fireplace quality feature is a bad choice since it has a lot of missing values and the top two categories are of similar count. When the top two categories (e.g., "Gd" and "TA") are close in count, theres no clear mode, and choosing one over the other (say "Gd") introduces bias.The first plot here shows that the missing values have values typically higher than good. The second plot for feature quality shows that since the graphs do not overlap the values are distorted. This might be because now the imputed kde plot is more influenced by Good category.

Implication: This suggests the missing entries might actually belong to higher-quality fireplaces, or at least not Gd.That contradicts what mode imputation does (which assumes the missing ones are like the majority  possibly Gd).

4. Another option is to just create a missing category(when missing rows >10%).This retains the missingness information, which can be meaningful  especially if missing values aren't random. However,creating 'Missing' category fails to capture meaningful patterns in the data and instead acts as noise, especially when SalePrice is strongly influenced by the actual fireplace quality level."
