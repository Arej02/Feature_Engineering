```python
import pandas as pd
```

```python
placement=pd.read_csv("placement.csv")
placement.sample(3)
```

# 1) Using the Z-score method:

```python
import matplotlib.pylab as plt
plt.style.use('fivethirtyeight')

fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(15,6))

ax[0].hist(placement['cgpa'])
ax[0].set_title("Histogram of CGPA")

ax[1].hist(placement['placement_exam_marks'],color='red')
ax[1].set_title("Histogram of Placement Exam")

fig.show()
```

```python
print(f"Skewness of CGPA is:{placement['cgpa'].skew()}")
print(f"Skewness of placement exams is:{placement['placement_exam_marks'].skew()}")
```

```python
mean=placement['cgpa'].describe()
```

```python
mean=placement['cgpa'].mean()
std=placement['cgpa'].std()
print(mean,std)
```

```python
upper_limit=mean+3*std
lower_limit=mean-3*std
print(upper_limit,lower_limit)
```

```python
placement[(placement['cgpa']>upper_limit) | (placement['cgpa']<lower_limit)]
```

```python
#Trimming:
trimmed_df=placement[(placement['cgpa']<upper_limit) & (placement['cgpa']>lower_limit)]
trimmed_df.describe()
```

```python
trimmed_df.shape
```

```python
placement.describe()
```

```python
#Capping:
capped_df=placement['cgpa'].clip(lower=lower_limit,upper=upper_limit)
capped_df.describe()
```

```python
capped_df.shape
```

# 2. Using the IQR method

```python
placement.describe()
```

```python
#Since placement is not normally distributed and is a skewed feature we can use the IQR method:
percentile25=placement['placement_exam_marks'].quantile(0.25)
percentile75=placement['placement_exam_marks'].quantile(0.75)
IQR=percentile75-percentile25
```

```python
upper_limit2=percentile75+1.5*IQR
lower_limit2=percentile25-1.5*IQR
print(upper_limit2,lower_limit2)
```

```python
placement[placement['placement_exam_marks']>84.5]
```

```python
trimmed_df2=placement[placement['placement_exam_marks']<84.5]
trimmed_df2.shape
```

```python
import seaborn as sns
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,10))

ax[0,0].set_title("Before Trimming")
ax[0,1].set_title("After Trimming")
sns.histplot(placement['placement_exam_marks'],ax=ax[0,0],kde=True,color='green')
sns.histplot(trimmed_df2['placement_exam_marks'],ax=ax[0,1],kde=True)

sns.boxplot(placement['placement_exam_marks'],ax=ax[1,0],color='green')
sns.boxplot(trimmed_df2['placement_exam_marks'],ax=ax[1,1])

fig.show()
```

```python
capped_df2=placement['placement_exam_marks'].clip(lower=lower_limit2,upper=upper_limit2)
capped_df2=pd.DataFrame(capped_df2)
capped_df2.describe()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,10))

ax[0,0].set_title("Before Capping")
ax[0,1].set_title("After Capping")
sns.histplot(placement['placement_exam_marks'],ax=ax[0,0],kde=True,color='green')
sns.histplot(capped_df2['placement_exam_marks'],ax=ax[0,1],kde=True)

sns.boxplot(placement['placement_exam_marks'],ax=ax[1,0],color='green')
sns.boxplot(capped_df2['placement_exam_marks'],ax=ax[1,1])

fig.show()
```

**Important Points**
1. It is easy to detect outliers, but what you to do with the outliers after identifying it is an issue. Do we want to keep it, or completely remove it depends on the problem statement. We Keep them (if they carry meaning), Remove them (if they are noise), Cap or transform them (to reduce their influence)
2. You can not always remove outliers. Say, what if you want to do anamoly detection.
3. Weight-based algorithm (Linear Regression, Logistic Regression, Adaboost) are affected by outliers as they depend on weights,whereas, tree-based algorithm might not be affected as much.
4. The first approach I have used is the z-score outlier detection and treatment. For any feature that follows normal distribution, we can either trim completely remove the rows) or cap (Replace extreme values with a max/min cap) the values.
5. The second appraoch I have used is IQR, for features that are not normally distributed (Skewed).Like the above, I have used two appraoches trimming and capping. Sometimes trimming might cause loss of a lot of rows so capping might be an better option.
