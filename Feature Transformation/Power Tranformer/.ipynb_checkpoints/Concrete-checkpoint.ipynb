```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
plt.style.use('fivethirtyeight')
import seaborn as sns
```

```python
df=pd.read_csv("concrete_data.csv")
df.shape
```

```python
df.sample(3)
```

```python
df.info()
```

```python
df.isnull().mean()*100
```

```python
X=df.drop(columns=['Strength'])
Y=df['Strength']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
Y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

print(f"The R2 score is:{r2_score(Y_test,Y_pred)}")
```

```python
lr_scores=cross_val_score(lr,X_train,Y_train,cv=5,scoring='r2')
print(f"Mean R2 Score is:{lr_scores.mean():.3f}(+/-{lr_scores.std():.3f})")
```

```python
numerical_cols=[cols for cols in df.select_dtypes(include=['float64','int64']).columns if cols!='Strength']
```

```python
import statsmodels.api as sm
def univariate_plot(column):
    fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(10,10))
    
    sns.histplot(data=df,x=column,kde=True,color='blue',ax=ax[0])
    sm.qqplot(df[column],ax=ax[1],line='s')

     #Summary statistics:
    mean=df[column].mean()
    median=df[column].median()
    std=df[column].std()
    skew=df[column].skew()

    stats_text=(
        f"Mean:{round(mean,2)}\n"
        f"Median:{round(median,2)}\n"
        f"Std:{round(std,2)}\n"
        f"Skewness:{round(skew,2)}\n"
     )

    ax[0].text(
        0.98,0.98,
        stats_text,
        ha='right',
        va='top',
        transform=ax[0].transAxes,
        bbox=dict(facecolor='white', alpha=0.7,edgecolor='black',linewidth=1.5),
        fontsize=10
    )

    plt.show()
```

```python
for values in numerical_cols:
    univariate_plot(values)
```

```python
from sklearn.preprocessing import PowerTransformer

pt=PowerTransformer()

X_train_transformed=pt.fit_transform(X_train)
X_test_transformed=pt.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
```

```python
lr.fit(X_train_transformed,Y_train)
```

```python
Y_pred=lr.predict(X_test_transformed)
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score

print(f"The R2 score is:{r2_score(Y_test,Y_pred)}")
```

```python
lr_scores=cross_val_score(lr,X_train_transformed,Y_train,cv=5,scoring='r2')
print(f"Mean R2 Score is:{lr_scores.mean():.3f}(+/-{lr_scores.std():.3f})")
```
